# Interpolate SSTs generated using the MSMIP methods across the land sea masks to get a full SST fields for use as an AMIP type boundary forced data.

```python
import xarray as xr

```

```python
dir_in = '/glade/work/rneale/MSMIP/sst/e3smv2/'
#file_in = 'DOE.E3SMv2.TS.LatBlend_30.CGCM.1980-2014.nc'
file_in = 'DOE.E3SMv2.TS.LatBlend_30.AGCM_1dRandPatt.1980-2014.nc'
ds = xr.open_dataset(dir_in+file_in) 

```

```python
ds
```

```python

da = ds['TS_AGCM_1dRandPatt']
da.isel(time=0).plot()
```

```python
da = da.isel(time=0).interpolate_na(dim="lat", method="linear")
```
