```python
import numpy as np
import matplotlib.pyplot as mp
import matplotlib

""" ########################### """
""" ### Average bar chart ##### """
""" ########################### """
pubs_yr = [40,64,104,129,197,268,282,319,329,342,381,422,470]
year = np.arange(2009,2022)
print(year)
print(pubs_yr)

fig, ax = mp.subplots(1,3,figsize=(35,10))

ax[0].bar(year,pubs_yr)
ax[0].set_ylabel('# Submissions',fontsize=29)
ax[0].set_xlabel('Year',fontsize=29)
ax[0].set_title('GMD Submission Growth',fontsize=29)



te_current = [3,11,12,10,14,12,4,3,1]
te_12mnths = [4,3,5,7,5,5,7,3,8,6,1,3,2,8]
te_pub_c = np.arange(0,9)
te_pub_12m = np.arange(0,14)

print(te_pub_c)
print(te_current)


ax[1].bar(te_pub_12m,te_12mnths)
ax[1].set_ylabel('TEs',fontsize=29)
ax[1].set_xlabel('# Submissions',fontsize=29)
ax[1].set_title('TE Handling (12 months)',fontsize=29)


ax[2].bar(te_pub_c,te_current)
ax[2].set_ylabel('TEs',fontsize=29)
ax[2].set_xlabel('# Submissions',fontsize=29)
ax[2].set_title('TE Handling (current)',fontsize=25)

for label in (ax[0].get_xticklabels() + ax[0].get_yticklabels()):
        label.set_fontsize(20)
for label in (ax[1].get_xticklabels() + ax[1].get_yticklabels()):
        label.set_fontsize(20)
for label in (ax[2].get_xticklabels() + ax[2].get_yticklabels()):
        label.set_fontsize(20)

mp.savefig('gmd_te1.png', dpi=200)
```
