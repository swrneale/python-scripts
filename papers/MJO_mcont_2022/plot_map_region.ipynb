```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

import cartopy.crs as ccrs
from cartopy.feature import LAND

desired_proj = ccrs.PlateCarree()

fig = plt.figure(figsize=(10,10))
ax = plt.subplot(projection=desired_proj)

ax.set_global()
ax.set_extent([80, 170, -20, 20])


ax.add_patch(mpatches.Rectangle(xy=[90, -18], width=70, height=36,
                                    facecolor='blue',
                                    alpha=0.2,
                                    transform=ccrs.PlateCarree())
                 )
ax.gridlines()
ax.coastlines()
ax.add_feature(LAND,color='k')

plt.savefig('bar_plot_ave_region.png', dpi=120, bbox_inches='tight') 
plt.show()
```
