```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

# Create a sample xarray DataArray with longitude ranging from 0 to 360
lon_original = np.linspace(0, 360, 10)
data = np.random.rand(10)
da = xr.DataArray(data, dims='lon', coords={'lon': lon_original})

# Change the longitude values to start from 90W (270E)
da['lon'] = (da['lon'] + 180 + 90) % 360
da=da[::-2]
print(da['lon'])

# Plot the DataArray
plt.plot(da['lon'], da, label='Data')


# Customize x-axis labels for both east and west longitudes
plt.xticks(np.arange(da['lon'].min(), da['lon'].max(), 45))
plt.gca().set_xticklabels([f'{abs(tick)}°W' if tick < 180 else f'{tick-180}°E' for tick in plt.xticks()[0]])

# Set plot labels and title
plt.xlabel('Longitude')
plt.ylabel('Data')
plt.title('Data Plot with Correct East and West Longitude Labels')

# Add legend
plt.legend()

# Show the plot
plt.show()

```
